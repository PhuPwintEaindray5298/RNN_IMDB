```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#Sentences

sent = [    'the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'I am a good boy',
        'I am a good developer',
        'understand the meaning of words',
        'your videos are good']
```

```python
# Define the vocabulary size

voc_size = 10000
```

```python
# One hot Representation 
one_hot_rep = [one_hot(words,voc_size) for words in sent]
```

```python
one_hot_rep
```

```python
#Word Embedding representation
from tensorflow.keras.layers import Embedding,Input
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_len = 5
embedded_doc = pad_sequences(one_hot_rep,padding='pre',maxlen=sent_len)
print(embedded_doc)
```

```python
#Feature representation
dim = 10
```

```python
model = Sequential()
model.add(Input(shape=(sent_len,)))
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
